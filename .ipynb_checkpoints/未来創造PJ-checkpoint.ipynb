```python
# -*- coding: utf-8 -*-
import requests
import pprint
import json

#レストラン検索APIのURL
url = "https://api.gnavi.co.jp/ForeignRestSearchAPI/v3/"

#パラメータの設定
params={}
params["keyid"] = "7a5527507df2f79d5ce305da65cb6771"
params["areacode_s"] = "AREAS2107" # 新橋
params['lang'] = 'en'
params['religious_menu_options'] = 1
params["freeword"] = "fish"

#リクエスト結果
result_api = requests.get(url, params)
result_api = result_api.json() 
# print(result_api) # 整形せずにそのまま表示
#pprint.pprint(result_api) # 整形して表示

hit = len(result_api['rest'])
print('国・宗教別メニュー相談可の店が' + str(hit) + '件見つかりました．')

# ループで、ヒットした10件分の店名を表示させる
for i in range(hit):
    print()
    print(result_api['rest'][i]["name"]['name'])
    print(result_api['rest'][i]["name"]['name_sub'])
    print(result_api['rest'][i]['contacts']["address"])
    print(result_api['rest'][i]['business_hour'])
    print(result_api['rest'][i]['contacts']['tel'])
    print(result_api['rest'][i]["image_url"]['thumbnail'])
```
